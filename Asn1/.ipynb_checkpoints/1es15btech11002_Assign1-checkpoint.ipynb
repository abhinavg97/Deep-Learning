## 1b
### Difference between Tensor, storage and variable

Variables are just wrappers for the tensors so you can easily auto compute the gradients.

The Variable class is a wrapper over torch Tensors (nd arrays in torch) that supports nearly all operations defined on tensors

#### Tensor

A torch.Tensor is a multi-dimensional matrix containing elements of a single data type.

Each tensor has an associated torch.Storage, which holds its data. The tensor class provides multi-dimensional, strided view of a storage and defines numeric operations on it.

#### Storage 
A torch.Storage is a contiguous, one-dimensional array of a single data type.

Every torch.Tensor has a corresponding storage of the same data type.

#### Variable

Wraps a tensor and records the operations applied to it.

## 2a

```python
#!/usr/bin/env python
from __future__ import print_function
from itertools import count

import torch
import torch.autograd
import torch.nn.functional as F
import torch.nn as nn
from torch.autograd import Variable

POLY_DEGREE = 4
W_target = torch.randn(POLY_DEGREE, 1) * 2
b_target = torch.randn(1) * 2


def make_features(x):
    """Builds features i.e. a matrix with columns [x, x^2, x^3, x^4]."""
    x = x.unsqueeze(1)
    return torch.cat([x ** i for i in range(1, POLY_DEGREE+1)], 1)


def f(x):
    """Approximated function."""
    return x.mm(W_target) + b_target[0]


def poly_desc(W, b):
    """Creates a string description of a polynomial."""
    result = 'y = '
    for i, w in enumerate(W):
        result += '{:+.2f} x^{} '.format(w, len(W) - i)
    result += '{:+.2f}'.format(b[0])
    return result


def get_batch(batch_size=32):
    """Builds a batch i.e. (x, f(x)) pair."""
    random = torch.randn(batch_size)
    x = make_features(random)
    y = f(x)
    return Variable(x), Variable(y)


# Define model
fc = torch.nn.Linear(W_target.size(0), 1)
optimizer = torch.optim.SGD(fc.parameters(), lr=0.006, momentum=0.1)
criterion = nn.MSELoss()# Mean Squared Loss

for batch_idx in count(1):
    # Get data
    inputs, target = get_batch()

    # Reset gradients
    optimizer.zero_grad()
    
    inputs = inputs.type(torch.FloatTensor)
    target = target.type(torch.FloatTensor)
    # Forward pass
    output = fc.forward(inputs)
    
    loss = criterion(output, target)
    
    # Backward pass
    loss.backward()

    optimizer.step()
#     print (loss.data[0])  #uncomment this line to see effect of learning rate, as we increase the learning rate,loss blows up
  #  Stop criterion
    if loss.data[0] < 1e-3:
        break

print('Loss: {:.6f} after {} batches'.format(loss.data[0], batch_idx))
print('==> Learned function:\t' + poly_desc(fc.weight.data.view(-1), fc.bias.data))
print('==> Actual function:\t' + poly_desc(W_target.view(-1), b_target))
```

### Comments on effects of changing learning rate
#### As the learning rate is increased the loss blows up, meaning we have shooted past the optimum value
#### Decreasing the learning rate to 0.06 from 0.07 gets us to a convergence

## 2b

### Taking input 

```python


# Load CSV
import numpy as np
import torch.nn as nn
from torch.autograd import Variable
import torch

filename = 'qn2_data.csv'
raw_data = open(filename, 'rt')
data = np.loadtxt(raw_data, delimiter=",")
x_train = data[:,:2].astype(np.float32)
y_train = data[:,2:].astype(np.float32)
```

### Defining Linear regression model

```python
model = torch.nn.Linear(2,1)

criterion = nn.MSELoss() # Mean Squared Loss
l_rate = 0.0017    # this loss rate converges to a good solution fastest
optimiser = torch.optim.SGD(model.parameters(), lr = l_rate) #Stochastic Gradient Descent

epochs = 40000
```

```python
def get_batch(x_train, y_train):
    """Builds a batch i.e. (x, f(x)) pair."""
    batch_size = 6
    random = np.random.choice(range(x_train.shape[0]),batch_size , replace = False)
    xtemp = []
    ytemp = []
    for i in range(batch_size):
        xtemp.append(x_train[i])
        ytemp.append(y_train[i])
    xtemp = np.array(xtemp)
    ytemp = np.array(ytemp)
    xtemp = torch.from_numpy(xtemp)
    ytemp = torch.from_numpy(ytemp)

    return Variable(xtemp), Variable(ytemp)

for epoch in range(epochs):

    epoch +=1
    #increase the number of epochs by 1 every time
    inputs, labels = get_batch(x_train, y_train)

    optimiser.zero_grad()
    #forward to get predicted values
    outputs = model(inputs)
    loss = criterion(outputs, labels)
    loss.backward()# back props
    optimiser.step()# update the parameters
    print('epoch {}, loss {}'.format(epoch,loss.data[0]))
```

### Loading train and test data

```python
from __future__ import print_function

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.autograd import Variable
from torch.optim.lr_scheduler import StepLR
import matplotlib.pyplot as plt

batch_size = 64
test_batch_size = 1000
seed = 1
log_interval = 193

torch.manual_seed(seed)

train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=True, download=True,
                   transform=transforms.Compose([
                       transforms.ToTensor(),
                       transforms.Normalize((0.1307,), (0.3081,))
                   ])),
    batch_size=batch_size, shuffle=True)

test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=False, transform=transforms.Compose([
                       transforms.ToTensor(),
                       transforms.Normalize((0.1307,), (0.3081,))
                   ])),
    batch_size=test_batch_size, shuffle=True)
```

### Methods for training, testing and plotting

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(320, 50)
        self.fc2 = nn.Linear(50, 10)

    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2))
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))
        x = x.view(-1, 320)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, training=self.training)
        x = self.fc2(x)
        return F.log_softmax(x)
    


def train(epoch):
    model.train()
    train_loss = 0
    correct = 0

    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = Variable(data), Variable(target)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target, size_average = False)
        train_loss += loss.data[0]
        pred = output.data.max(1, keepdim = True)[1]
        correct += pred.eq(target.data.view_as(pred)).cpu().sum()
        loss.backward()
        optimizer.step()
        
        if batch_idx % log_interval == 0 :
            print('Train Epoch {}: [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.data[0]/len(data)))
        
    train_loss /= len(train_loader.dataset)    
    trainloss.append(train_loss)
    
    print('\nTrain set {}: Average loss: {:.4f}, Training classification error: Misclassified Samples {}/{} ({:.0f}%)\n'.format( 
        epoch, train_loss, len(train_loader.dataset) - correct, len(train_loader.dataset),
        100.  *  (1.0*(len(train_loader.dataset) - correct) / len(train_loader.dataset))) )
            

# This code is very similar to the train code provided

def test(epoch):
    # TODO: Test the model on the test-set and report the loss and accuracy.
    model.eval()
    test_loss = 0
    correct = 0
    for data, target in test_loader:
        data, target = Variable(data, volatile = True), Variable(target) #volatile true means the data is read from memory rather than register everytime
        output = model(data)
        loss = F.nll_loss(output, target, size_average = False) # sum up batch loss, by default it averages
        test_loss += loss.data[0]                               # summing up the test loss, we'll later average it
        pred = output.data.max(1, keepdim = True)[1]
        correct += pred.eq(target.data.view_as(pred)).cpu().sum()  # summing the number of instatnces predicted correctly by the model
        
    test_loss /= len(test_loader.dataset)  # taking the average of the test_loss
    testloss.append(test_loss)          
    print('\nTest set {}: Average loss: {:.4f}, Test Classification Error: Misclassified samples {}/{} ({:.0f}%)\n'.format(  # classification error
        epoch, test_loss, len(test_loader.dataset) - correct, len(test_loader.dataset),
        100. *   (1.0*(len(test_loader.dataset) - correct) / len(test_loader.dataset))) )
    
def graph():
    plt.title('Training Loss, Test loss vs epochs')
    plt.xlabel('No. of Epochs')
    plt.ylabel('Loss')
    plt.plot(no_epoch, trainloss, label = 'Train Loss')
    plt.plot(no_epoch, testloss, label = 'Test Loss')
    plt.legend(loc = 'best')
    plt.show()
```

### SGD 

```python
model = Net()

# Training settings

epochs = 4
lr = 0.001
momentum = 0.5
dampening = 0
weight_decay = 0.2
nesterov = True

optimizer = optim.SGD(model.parameters(), lr=lr, momentum = momentum, dampening = dampening,
                      weight_decay = weight_decay, nesterov = nesterov)

trainloss = []
testloss = []
no_epoch = []

scheduler = StepLR(optimizer, step_size=2, gamma=0.1) # dynamic learning rate, changes lr after step number of epochs
for epoch in range(1, epochs + 1):
    scheduler.step()
    no_epoch.append(epoch)
    train(epoch)
    test(epoch)
    
graph()
```

### Adagrad

```python
model = Net()

# Training settings

epochs = 4
lr = 0.01
lr_decay = 0  
weight_decay = 0


optimizer = optim.Adagrad(model.parameters(), lr=lr, lr_decay = lr_decay, weight_decay = weight_decay)

trainloss = []
testloss = []
no_epoch = []

scheduler = StepLR(optimizer, step_size=2, gamma=0.1)
for epoch in range(1, epochs + 1):
    scheduler.step()
    no_epoch.append(epoch)
    train(epoch)
    test(epoch)
    
graph()
```

```python
model = Net()

# optimizer = optim.LBFGS(model.parameters(), lr=1, max_iter=6, max_eval=None, 
#                         tolerance_grad=1e-05, tolerance_change=1e-09, 
#                         history_size=40, line_search_fn=None)

# Training settings

epochs = 4
lr = 0.016
weight_decay = 0
eps = 1e-08
betas = (0.9, 0.999)

optimizer = optim.Adam(model.parameters(), lr=lr, betas = betas, eps = eps, weight_decay = weight_decay)

trainloss = []
testloss = []
no_epoch = []

scheduler = StepLR(optimizer, step_size=2, gamma=0.1)

for epoch in range(1, epochs + 1):
    scheduler.step()
    no_epoch.append(epoch)
    train(epoch)
    test(epoch)
    
graph()
```

### Why the solution is "good"
